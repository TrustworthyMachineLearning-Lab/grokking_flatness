```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import math
import yaml
```

```python
# Please select the directory under default directory
dir_name = "version_14"
```

```python
with open(os.path.join(dir_name, 'hparams.yaml'), 'r') as f:
    hparams = yaml.load(f, Loader=yaml.SafeLoader)
f.close()
```

```python
# the path to store the accuracy and loss data.
store_train_plot_data = hparams["store_train_plot_data"]
```

```python
# read in data from csv/excel files.
df = pd.read_csv(os.path.join(dir_name, "metrics.csv"))
train_loss = df["train_loss"].tolist()
train_accuracy = df["train_accuracy"].tolist()
epoch = df["epoch"].tolist()
step = df["step"].tolist()
val_loss = df["val_loss"].tolist()
val_accuracy = df["val_accuracy"].tolist()
full_train_loss = df["full_train_loss"].tolist()
full_train_acc = df["full_train_acc"].tolist()
```

```python
len(train_loss), len(train_accuracy), len(epoch), len(step), len(val_loss), len(val_accuracy), len(full_train_loss), len(full_train_acc)
```

```python
train_loss_step_dict = dict(zip(step, train_loss))
train_accuracy_step_dict = dict(zip(step,train_accuracy))
val_loss_step_dict = dict(zip(step, val_loss))
val_accuracy_step_dict = dict(zip(step, val_accuracy))
full_train_loss_step_dict = dict(zip(step, full_train_loss))
full_train_acc_step_dict = dict(zip(step, full_train_acc))
```

```python
# remove nan values 
train_loss_x, train_loss_y = [], []
for key, value in train_loss_step_dict.items():
    if math.isnan(value):
        continue
    else:
        train_loss_x.append(key)
        train_loss_y.append(value)
```

```python
len(train_loss_x), len(train_loss_y),train_loss_x[-1]
```

```python
# remove nan values
val_loss_x, val_loss_y = [], []
for key, value in val_loss_step_dict.items():
    if math.isnan(value):
        continue
    else:
        val_loss_x.append(key)
        val_loss_y.append(value)
```

```python
len(val_loss_x), len(val_loss_y), val_loss_x[-1]
```

```python
# remove the nan values
train_acc_x, train_acc_y = [], []
for key, value in train_accuracy_step_dict.items():
    if math.isnan(value):
        continue
    else:
        train_acc_x.append(key)
        train_acc_y.append(value)
```

```python
len(train_acc_x), len(train_acc_y)
```

```python
# remove the nan values
val_acc_x, val_acc_y = [], []
for key, value in val_accuracy_step_dict.items():
    if math.isnan(value):
        continue
    else:
        val_acc_x.append(key)
        val_acc_y.append(value)
```

```python
len(val_acc_x), len(val_acc_y)
```

```python
# store the acc data to the hessian/sharpness/weight norm/sharpness2 dirs
np.save(os.path.join(store_train_plot_data, "train_acc_x"), np.array(train_acc_x))
np.save(os.path.join(store_train_plot_data, "train_acc_y"), np.array(train_acc_y))
np.save(os.path.join(store_train_plot_data, "val_acc_x"), np.array(val_acc_x))
np.save(os.path.join(store_train_plot_data, "val_acc_y"), np.array(val_acc_y))
```
