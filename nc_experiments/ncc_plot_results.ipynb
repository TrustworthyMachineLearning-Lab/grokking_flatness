```python
import matplotlib.pyplot as plt 
import numpy as np
import os
import collections
```

```python
os.getcwd().split("/")[-1].split("TSIZE0.0_")[-1]
```

```python
dir_name = "."
dirs = os.listdir(".")
```

```python
# load data
file_name_sharpness = []
file_name_sharpness2 = []
file_name_train_loss = []
file_name_train_acc = []
file_name_val_loss = []
file_name_val_acc = []
file_name_weights_norm = []
file_name_cluster_value = []
file_name_mean_value = []
file_name_var_classcluster = []
for item in dirs:
    if "sharpness_" in item:
        file_name_sharpness.append(item)
    elif "sharpness2_" in item:
        file_name_sharpness2.append(item)
    elif "train_acc_" in item:
        file_name_train_acc.append(item)
    elif "val_acc_" in item:
        file_name_val_acc.append(item)
    elif "train_loss_" in item:
        file_name_train_loss.append(item)
    elif "val_loss_" in item:
        file_name_val_loss.append(item)
    elif "weights_norm_" in item:
        file_name_weights_norm.append(item)
    elif "cluster_value" in item:
        file_name_cluster_value.append(item)
    elif "mean_value" in item:
        file_name_mean_value.append(item)
    elif "var_classcluster" in item:
        file_name_var_classcluster.append(item)
    
```

```python
len(file_name_sharpness), len(file_name_sharpness2), len(file_name_weights_norm), len(file_name_cluster_value)
```

```python
len(file_name_train_acc), len(file_name_train_loss), len(file_name_val_acc), len(file_name_val_loss)
```

```python
sharpness_dict = {}
for item in file_name_sharpness:
    temp_int = item.split("_")[-1].split(".")[0]
    sharpness_dict[int(temp_int)] = item

sharpness_dict = collections.OrderedDict(sorted(sharpness_dict.items()))
sharpness_array = []
sharpness_step_list = []
for key, value in sharpness_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    sharpness_step_list.append(key)
    sharpness_array.append(temp_data)
sharpness_array = np.array(sharpness_array)
```

```python
sharpness2_dict = {}
for item in file_name_sharpness2:
    temp_int = item.split("_")[-1].split(".")[0]
    sharpness2_dict[int(temp_int)] = item

sharpness2_dict = collections.OrderedDict(sorted(sharpness2_dict.items()))
sharpness2_array = []
sharpness2_step_list = []
for key, value in sharpness2_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    sharpness2_step_list.append(key)
    sharpness2_array.append(temp_data)
sharpness2_array = np.array(sharpness2_array)
```

```python
train_loss_dict = {}
for item in file_name_train_loss:
    temp_int = item.split("_")[-1].split(".")[0]
    train_loss_dict[int(temp_int)] = item

train_loss_dict= collections.OrderedDict(sorted(train_loss_dict.items()))
train_loss_array = []
train_loss_step_list = []
for key, value in train_loss_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    train_loss_step_list.append(key)
    train_loss_array.append(temp_data)
train_loss_array = np.array(train_loss_array)
```

```python
val_loss_dict = {}
for item in file_name_val_loss:
    temp_int = item.split("_")[-1].split(".")[0]
    val_loss_dict[int(temp_int)] = item

val_loss_dict= collections.OrderedDict(sorted(val_loss_dict.items()))
val_loss_array = []
val_loss_step_list = []
for key, value in val_loss_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    val_loss_step_list.append(key)
    val_loss_array.append(temp_data)
val_loss_array = np.array(val_loss_array)
```

```python
val_acc_dict = {}
for item in file_name_val_acc:
    temp_int = item.split("_")[-1].split(".")[0]
    val_acc_dict[int(temp_int)] = item

val_acc_dict= collections.OrderedDict(sorted(val_acc_dict.items()))
val_acc_array = []
val_acc_step_list = []
for key, value in val_acc_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    val_acc_step_list.append(key)
    val_acc_array.append(temp_data)
val_acc_array = np.array(val_acc_array)
```

```python
train_acc_dict = {}
for item in file_name_train_acc:
    temp_int = item.split("_")[-1].split(".")[0]
    train_acc_dict[int(temp_int)] = item

train_acc_dict= collections.OrderedDict(sorted(train_acc_dict.items()))
train_acc_array = []
train_acc_step_list = []
for key, value in train_acc_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    train_acc_step_list.append(key)
    train_acc_array.append(temp_data)
train_acc_array = np.array(train_acc_array)
```

```python
train_cluster_dict = {}
for item in file_name_cluster_value:
    temp_int = item.split("_")[-1].split(".")[0]
    train_cluster_dict[int(temp_int)] = item

train_cluster_dict= collections.OrderedDict(sorted(train_cluster_dict.items()))
train_cluster_array = []
train_cluster_step_list = []
for key, value in train_cluster_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    train_cluster_step_list.append(key)
    train_cluster_array.append(temp_data)
train_cluster_array = np.array(train_cluster_array)
```

```python
weightnorm_dict = {}
for item in file_name_weights_norm:
    temp_int = item.split("_")[-1].split(".")[0]
    weightnorm_dict[int(temp_int)] = item

weightnorm_dict= collections.OrderedDict(sorted(weightnorm_dict.items()))
weightnorm_array = []
weightnorm_step_list = []
for key, value in weightnorm_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    weightnorm_step_list.append(key)
    weightnorm_array.append(temp_data)
weightnorm_array = np.array(weightnorm_array)
```

```python
mean_value_dict = {}
for item in file_name_mean_value:
    temp_int = item.split("_")[-1].split(".")[0]
    mean_value_dict[int(temp_int)] = item

mean_value_dict= collections.OrderedDict(sorted(mean_value_dict.items()))
mean_value_array = []
mean_value_step_list = []
for key, value in mean_value_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    mean_value_step_list.append(key)
    mean_value_array.append(np.mean(np.linalg.norm((temp_data[:,0] - temp_data[:,1]), axis=1)))
mean_value_array = np.array(mean_value_array)

var_classcluster_dict = {}
for item in file_name_var_classcluster:
    temp_int = item.split("_")[-1].split(".")[0]
    var_classcluster_dict[int(temp_int)] = item

var_classcluster_dict= collections.OrderedDict(sorted(var_classcluster_dict.items()))
var_classcluster_array = []
var_classcluster_step_list = []
for key, value in var_classcluster_dict.items():
    temp_data = np.load(os.path.join(dir_name, value))
    var_classcluster_step_list.append(key)
    var_classcluster_array.append(np.mean(temp_data[:, 1] + temp_data[:, 3]))
var_classcluster_array = np.array(var_classcluster_array)

```

```python
t = os.getcwd().split("/")[-1].split("TSIZE0.0_")[-1].split("_")
t1 = "_".join(t[:4])
t2 = "_".join(t[4:])
```

```python
fig, ax = plt.subplots(8,1, figsize=(4,15))
fig.suptitle("{} \n {}".format(t1, t2))
# fig.suptitle('All in all\n')

# set colors
cmap = plt.get_cmap('viridis')
positions = np.linspace(0, 0.9, 10)
colors = [cmap(pos) for pos in positions]

ax[0].plot(train_acc_step_list, train_acc_array, color=colors[0], label="train acc")
ax[0].plot(val_acc_step_list, val_acc_array, color=colors[1], label="val acc", linestyle='-.')
ax[0].legend()
ax[1].plot(train_loss_step_list, train_loss_array, color=colors[2], label="train loss")
ax[1].plot(val_loss_step_list, val_loss_array, color=colors[3], label="val loss", linestyle='-.')
ax[1].legend()
ax[2].plot(sharpness_step_list, sharpness_array, color=colors[4], label="train sharpness")
ax[3].plot(sharpness2_step_list, sharpness2_array, color=colors[5], label="train RF")
ax[4].plot(weightnorm_step_list, weightnorm_array, color=colors[6], label="weightnorm")
ax[5].plot(train_cluster_step_list, train_cluster_array, color=colors[7], label="cluster")
ax[6].plot(mean_value_step_list, mean_value_array, color=colors[8], label="mean distance")
ax[7].plot(var_classcluster_step_list, var_classcluster_array, color=colors[9], label="var values")


ax[0].set_xlabel("Training Epochs")
ax[1].set_xlabel("Training Epochs")
ax[2].set_xlabel("Training Epochs")
ax[3].set_xlabel("Training Epochs")
ax[4].set_xlabel("Training Epochs")
ax[5].set_xlabel("Training Epochs")
ax[6].set_xlabel("Training Epochs")


ax[0].set_ylabel("Accuracy")
ax[1].set_ylabel("Loss")
ax[2].set_ylabel("Sharpness")
ax[3].set_ylabel("Relative Flatness")
ax[4].set_ylabel("Weights Norm")
ax[5].set_ylabel("Cluster")
ax[6].set_ylabel("Cluster(mean distance)")
ax[7].set_ylabel("Cluster(var)")


ax[0].spines[['right', 'top']].set_visible(False)
ax[1].spines[['right', 'top']].set_visible(False)
ax[2].spines[['right', 'top']].set_visible(False)
ax[3].spines[['right', 'top']].set_visible(False)
ax[4].spines[['right', 'top']].set_visible(False)
ax[5].spines[['right', 'top']].set_visible(False)
ax[6].spines[['right', 'top']].set_visible(False)
ax[7].spines[['right', 'top']].set_visible(False)


for item in ax.flat:
    item.grid(True, which='both', linestyle='--', linewidth=0.5, color='gray')

fig.tight_layout()
plt.savefig('{}.pdf'.format(dir_name), dpi=300, bbox_inches='tight')
```
